```python
df = df[df['Year'].isin([2019, 2020, 2021])]
df
```

```python
# Adding household zie
# count the number of occurrences of each serial number
Size = df.groupby('SERIALNO').size().reset_index(name='HHSize')

# join the 'HHSize' column to the original DataFrame on 'SERIALNO'
df = pd.merge(df, Size, on='SERIALNO', how='left')
```

```python
# Adding Children Info 
HHChild = df.copy()

# select the relevant columns
HHChild = HHChild[['SERIALNO', 'RELSHIPP', 'Age']]

# create a new column 'IsChild' and initialize it with NA
HHChild['IsChild'] = np.nan

# set 'IsChild' to 1 if 'Age' is less than 18 and 'RELSHIPP' is not 20, 21 or 23; otherwise set it to 0
HHChild.loc[(HHChild['Age'] < 18) & (~HHChild['RELSHIPP'].isin([20, 21, 23])), 'IsChild'] = 1
HHChild['IsChild'].fillna(value=0, inplace=True)

# group the data by 'SERIALNO' and count the number of children present
ChildPresent = HHChild.groupby('SERIALNO', as_index=False).agg({'IsChild': 'sum'})
ChildPresent['ChildPresent'] = np.where(ChildPresent['IsChild'] >= 1, 1, 0)

# join the 'ChildPresent' column to the original DataFrame on 'SERIALNO'
df = pd.merge(df, ChildPresent[['SERIALNO', 'ChildPresent']], on='SERIALNO', how='right')

df
```

```python
# Adjust income 
df.loc[df['Year'] == 2019, 'PINCP'] *= 344.416/299.433
df.loc[df['Year'] == 2020, 'PINCP'] *= 344.416/303.932
df.loc[df['Year'] == 2021, 'PINCP'] *= 344.416/319.761
```

```python
# Labor Force
df['LF'] = np.where((df['Age'] >= 16) & (df['Age'] <= 70) & (df['ESR'] != 6), 1, 0)
```

```python
# SEX
df['sex'] = np.where((df['Gender'] == 'M'), 1, 0)
```

```python
# Adding the income categories
df['IncomeCat_PINCP'] = np.where(df['PINCP'] == 9999999, 0,
                   np.where(df['PINCP'] < 15000, 1,
                   np.where((df['PINCP'] >= 15000) & (df['PINCP'] < 30000), 2,
                   np.where((df['PINCP'] >= 30000) & (df['PINCP'] < 45000), 3,
                   np.where((df['PINCP'] >= 45000) & (df['PINCP'] < 60000), 4,
                   np.where((df['PINCP'] >= 60000) & (df['PINCP'] < 75000), 5,
                   np.where((df['PINCP'] >= 75000) & (df['PINCP'] < 100000), 6,
                   np.where((df['PINCP'] >= 100000) & (df['PINCP'] < 125000), 7,
                   np.where((df['PINCP'] >= 125000) & (df['PINCP'] < 150000), 8,
                   np.where((df['PINCP'] >= 150000) & (df['PINCP'] < 200000), 9,
                   np.where((df['PINCP'] >= 200000) & (df['PINCP'] < 300000), 10,
                   np.where((df['PINCP'] >= 300000) & (df['PINCP'] < 500000), 11,
                   np.where((df['PINCP'] >= 500000) & (df['PINCP'] < 700000), 12,
                   np.where((df['PINCP'] >= 700000) & (df['PINCP'] < 900000), 13, 14)
                   )))))))))))))
```

```python
# Cleaning for 2019 
cleaned_2019 = copy.deepcopy(df)
cleaned_2019 = cleaned_2019[cleaned_2019['Year'] == 2019]
cleaned_2019 =  cleaned_2019.drop_duplicates(subset='SERIALNO', keep='first')

# Filtering out GQ
cleaned_2019 = cleaned_2019[cleaned_2019['TYPEHUGQ'] == 1]

# Turn MIL into categorical 
dummies = pd.get_dummies(cleaned_2019['MIL'], prefix='MIL')
cleaned_2019 = pd.concat([cleaned_2019, dummies], axis=1)

# ESR into categorical 
# Turn MIL into categorical 
dummies = pd.get_dummies(cleaned_2019['ESR'], prefix='ESR')
cleaned_2019 = pd.concat([cleaned_2019, dummies], axis=1)

# Turn COW (character of work) into categorical 
dummies = pd.get_dummies(cleaned_2019['COW'], prefix='COW')
cleaned_2019 = pd.concat([cleaned_2019, dummies], axis=1)

# Turn HISP into categorical 
dummies = pd.get_dummies(cleaned_2019['HISP'], prefix='HISP')
cleaned_2019 = pd.concat([cleaned_2019, dummies], axis=1)

# Turn HISP into categorical 
dummies = pd.get_dummies(cleaned_2019['RAC1P'], prefix='RAC1P')
cleaned_2019 = pd.concat([cleaned_2019, dummies], axis=1)

# Turn HISP into categorical 
dummies = pd.get_dummies(cleaned_2019['RaceCode'], prefix='RaceCode')
cleaned_2019 = pd.concat([cleaned_2019, dummies], axis=1)

cleaned_2019
cleaned_2019
```

```python
import statsmodels.api as sm
import pandas as pd
df = copy.deepcopy(smaller_set_of_variables)

# specify dependent variable
y = df['IncomeCat_PINCP']

# specify independent variables
X = df[['HHSize', 'AGEP', 'sex',
       'RaceCode_1', 'RaceCode_2', 'RaceCode_3', 'RaceCode_4', 'RaceCode_5',
       'RaceCode_6', 'RaceCode_7', 'RaceCode_8', 'RaceCode_9', 'LF']]

# specify weight variable
weights = df['WGTP']

# fit weighted multinomial logistic regression model
model = sm.MNLogit(y, sm.add_constant(X)).fit(method='bfgs', maxiter=100, w=weights)

# print summary of results
model.summary()
```
