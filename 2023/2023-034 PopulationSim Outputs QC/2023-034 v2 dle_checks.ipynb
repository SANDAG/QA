## NAICS data vs. Econ REMI data
For information only

```python
# Based on xwalk_PUMS_jobs.xlsx file from ENF

def PUMS_group(df):
    if df['NAICS2'] in ('92'):
        return 'job_1'
    elif df['NAICS2'] in ('MIL'):
        return 'job_2'
    elif df['NAICS2'] in ('11', '21'):
        return 'job_3'
    elif df['NAICS2'] in ('51', '54', '56'):
        return 'job_4'
    elif df['NAICS2'] in ('52', '53', '55'):
        return 'job_5'
    elif df['NAICS2'] in ('61'):
        return 'job_6'
    elif df['NAICS2'] in ('62'):
        return 'job_7'
    elif df['NAICS2'] in ('44', '45', '4M'):
        return 'job_8'
    elif df['NAICS2'] in ('48', '49', '23'):
        return 'job_9'
    elif df['NAICS2'] in ('22', '31', '32', '33', '42', '3M'):
        return 'job_10'
    elif df['NAICS2'] in ('71'):
        return 'job_11'
    elif df['NAICS2'] in ('721'):
        return 'job_12'
    elif df['NAICS2'] in ('722'):
        return 'job_13'
    elif df['NAICS2'] in ('81'):
        return 'job_14'
    else:
        return df['NAICS2']
    
synthetic_persons['xwalk_QC'] = synthetic_persons.apply(PUMS_group, axis='columns')
synthetic_persons
```

## HH Income/COW consistency check

Consistency check:  
Households with 0 income should not have workers, excluding COW classes 6 and 8

```python
# Method 1: based on WIF
hhinc_wif = synthetic_hh[(synthetic_hh['WIF'] > 0) & (synthetic_hh['HHADJINC'] == 0)]

hhinc_wif.to_excel(path + 'Results/hhinc0_worker_1.xlsx', index='False')
hhinc_wif
```

```python
# Method 2: based on COW

# Associate person characteristics with HH income
persons_hhincome = pd.merge(synthetic_persons, synthetic_hh, on='household_id')
persons_hhincome
```

```python
# Consistency check:
# Households with 0 income should not have workers, excluding COW classes 6 and 8
#   - 6: Self-employed in own not incorporated business, professional practice, or farm
#   - 8: Working without pay in family business or farm

hhinc_cow = persons_hhincome[(persons_hhincome['HHADJINC'] == 0) & 
                             (persons_hhincome['WKHP'] > 0) &
                             (~persons_hhincome['COW'].isin([0, 6, 8]))]
hhinc_cow = hhinc_cow[['household_id', 'SPORDER', 'AGEP', 'SEX', 'ESR', 'COW', 'WKHP', 'HHADJINC']]
hhinc_cow
```

```python
hhinc_cow.to_excel(path + 'Results/hhinc0_worker.xlsx', index='False')
```

## Ratio of workers

- Ratio of total workers to total employment
- Ratio of total part-time workers to total employment

```python
emp_tot = mgra_based['emp_tot'].sum()
emp_tot
```

```python
# Classify Part-Time/Full-Time workers based on WKHP (hours worked)
workers = synthetic_persons[:]
workers.loc[workers.ESR.isin([1,2,4,5]) & (workers.WKHP < 40) & (workers.WKHP > 0), 'work_status'] = 'PT'
workers.loc[workers.ESR.isin([1,2,4,5]) & (workers.WKHP >= 40), 'work_status'] = 'FT'
workers.loc[workers.ESR.isin([1,2,4,5]), 'isWorker'] = 1
workers
```

```python
workers_sum = workers.groupby(['work_status'])['isWorker'].agg(['count'])
worker_ratio = workers_sum.T # Transpose table

# Ratio of total workers to total employment
worker_ratio['TotalWorker_ratio'] = (worker_ratio['FT'] + worker_ratio['PT'])/emp_tot

# Ratio of PT to total workers
worker_ratio['PT_ratio'] = worker_ratio['PT']/(worker_ratio['PT'] + worker_ratio['FT'])

worker_ratio
```
