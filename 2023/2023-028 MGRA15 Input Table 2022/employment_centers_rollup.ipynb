```python
mgra_level_data = pd.read_csv(r'T:\socioec\Current_Projects\SR15\S0\version4\abm_csv\mgra15_based_input_2022_01.csv')
mgra_level_data = mgra_level_data.drop(['taz', 'LUZ'], axis=1)
mgra_level_data
```

```python
employment_centers_data = pd.read_excel(r'C:\Users\cra\San Diego Association of Governments\SANDAG QA QC - Documents\Projects\2023\2023-028 MGRA15 Input Table 2022\2023-028-02\Results\00_MASTER_MGRA15_to_EC2_XRef_2023_01_31.xlsx', sheet_name='MGRA-EC XRef 2023_01_24', header=1)
employment_centers_data = employment_centers_data[['MGRA15', 'EC_Name']]
employment_centers_data = employment_centers_data.dropna().reset_index(drop=True)
employment_centers_data
```

```python
output = mgra_level_data.merge(employment_centers_data, how='left', right_on='MGRA15', left_on='mgra')
output = output.drop(['mgra', 'MGRA15'], axis=1)
output = output.groupby('EC_Name').sum()
output['hhs'] = output['hhp']/output['hh']
output.to_excel('mgra_based_input_employment_centers_update.xlsx')
```
