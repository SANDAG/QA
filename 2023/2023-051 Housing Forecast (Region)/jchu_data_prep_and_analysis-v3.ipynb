```python
HS_calcs = pd.read_csv(r'T:\socioec\SR15 Models\Region Level Housing\Region Forecast Component Vac Model 5n2_09262023.csv')
HS_calcs
```

# 2.a. hh = owner + renter

```python
(HS_calcs['hh'] == HS_calcs['owner']+HS_calcs['renter']).sum()
```

# 2.b.	(2023-2060) NR_Total = previous year NR_Total + current year NR_growth

```python
HS_calcs_2023_2060 = HS_calcs[HS_calcs['year']>=2023].reset_index(drop=True)
HS_calcs_2023_2060
```

```python
# check the growth 
((round(HS_calcs_2023_2060['HS_chg'] * .044867).iloc[:-1]).reset_index(drop=True) == \
                 HS_calcs_2023_2060['NR_growth'].iloc[1:].reset_index(drop=True)).sum()
```

```python
(HS_calcs_2023_2060['NR_total'] == \
                 HS_calcs[HS_calcs['year']>=2022]['NR_total'].iloc[:-1].reset_index(drop=True) + \
                 HS_calcs_2023_2060['NR_growth'].reset_index(drop=True)).sum()
```

# 2.c.	(2023-2060) Total_HS = owner_hs + renter_hs + NR_total

```python
(HS_calcs_2023_2060['Total_HS'] == HS_calcs_2023_2060['owner_hs'] + HS_calcs_2023_2060['renter_hs'] + HS_calcs_2023_2060['NR_total']).sum()
```

# 2.d.	Vacancy = (Total_HS  hh) / Total_HS

```python
(round(HS_calcs['Vacancy'],6) == round((HS_calcs['Total_HS']-HS_calcs['hh'])/HS_calcs['Total_HS'], 6)).sum()
```
