# SM-6 Physical Activity

```python
conn = pyodbc.connect('Driver={ODBC Driver 17 for SQL Server};'
                    'Server=DDAMWSQL16.sandag.org;'
                    'Database=abm_14_2_0_reporting;'
                    'Trusted_Connection=yes;')

final_dict = {'Scenario_id':list(), 'Total_Time':list(), 'Percent':list()}


scenario_id = 767

def query_and_add_to_dict(final_dict, scenario_id):
  query =f'''
  SELECT measure,
    CASE
      WHEN measure LIKE '%Percentage%' THEN ROUND(value * 100, 1)
      ELSE ROUND(value, 2)
    END AS 'new_val'
  FROM [abm_14_2_0_reporting].[rp_2021].[results]
  WHERE scenario_id = {scenario_id} AND measure like 'SM-6%' AND metric = 'Total'
  '''

  sql_data =  pd.read_sql_query(query, conn)


  # add the new row to the dataframe
  #index_order = [2, 3, 4, 12, 5, 8, 9, 10, 13, 11]

  result = sql_data#.loc[index_order]
  final_dict['Scenario_id'].append(scenario_id)
  final_dict['Total_Time'].append(sql_data['new_val'][1])
  final_dict['Percent'].append(sql_data['new_val'][0])

  return final_dict

print(query_and_add_to_dict(final_dict, scenario_id))
result
```

```python
scenario_ids = [767, 759, 764, 773, 787, 789, 790, 780, 788, 791, 775, 777, 782, 784, 786, 785, 759, 766, 768]
final_dict = {'Scenario_id':list(), 'Total_Time':list(), 'Percent':list()}

for scenario_id in scenario_ids:
    query_and_add_to_dict(final_dict, scenario_id)

output = pd.DataFrame(final_dict).set_index('Scenario_id').T
output.to_csv('output.csv.')
```
