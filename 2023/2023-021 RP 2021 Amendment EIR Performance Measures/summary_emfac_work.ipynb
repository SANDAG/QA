```python
import pandas as pd
import os

def extract_annual_excel_columns(folder_path):
    """Extract specific columns from all Excel files containing 'annual' in file name within a folder
    
    Args:
        folder_path (str): The path to the folder containing the Excel files
    
    Returns:
        pd.DataFrame: A DataFrame containing the extracted columns from all Excel files
    """
    
    # Get a list of all files in the folder
    file_list = os.listdir(folder_path)
    
    # Filter by files containing 'annual' in the name
    annual_files = [file for file in file_list if 'annual' in file.lower() and file.endswith('.xlsx')]
    
    # Initialize an empty DataFrame to store the extracted columns
    extracted_df = pd.DataFrame()
    
    # Loop over each annual Excel file and extract the specified columns
    for file in annual_files:
        col1 = file.split('-')[2][:4] + '-' + file.split('-')[3]
        columns = pd.MultiIndex.from_tuples([(col1, 'CO2_TOTEX'), (col1, 'PM2_5_TOTEX'), (col1, 'PM10_TOTAL'), (col1, 'Fuel_GAS'), (col1, 'Fuel_DSL')])
        working_df = pd.DataFrame(columns=columns)
        
        # Read in the Excel file
        excel_path = os.path.join(folder_path, file)
        #df = pd.read_excel(excel_path)
        df = pd.read_excel(excel_path, sheet_name='Total SANDAG')
        df = df[df['Veh_Tech'] == ' All Vehicles']
        # Extract the specific columns of interest
        working_df[(col1, 'CO2_TOTEX')] = df['CO2_TOTEX']
        working_df[(col1, 'PM2_5_TOTEX')] = df['PM2_5_TOTEX']
        working_df[(col1, 'PM10_TOTAL')] = df['PM10_TOTAL']
        working_df[(col1, 'Fuel_GAS')] = df['Fuel_GAS']
        working_df[(col1, 'Fuel_DSL')] = df['Fuel_DSL']

        
        # Append the extracted columns to the main DataFrame
        extracted_df = pd.concat([extracted_df, working_df], axis=1)
        
    return extracted_df
```

```python
import pandas as pd
import os

def extract_annual_excel_columns(folder_path):
    """Extract specific columns from all Excel files containing 'annual' in file name within a folder
    
    Args:
        folder_path (str): The path to the folder containing the Excel files
    
    Returns:
        pd.DataFrame: A DataFrame containing the extracted columns from all Excel files
    """
    
    # Get a list of all files in the folder
    file_list = os.listdir(folder_path)
    
    # Filter by files containing 'annual' in the name
    annual_files = [file for file in file_list if 'summer' in file.lower() and file.endswith('.xlsx')]
    
    # Initialize an empty DataFrame to store the extracted columns
    extracted_df = pd.DataFrame()
    
    # Loop over each annual Excel file and extract the specified columns
    for file in annual_files:
        col1 = file.split('-')[2][:4] + '-' + file.split('-')[3]
        columns = pd.MultiIndex.from_tuples([(col1, 'ROG_TOTAL'), (col1, 'NOx_TOTEX')])
        working_df = pd.DataFrame(columns=columns)
        
        # Read in the Excel file
        excel_path = os.path.join(folder_path, file)
        #df = pd.read_excel(excel_path)
        df = pd.read_excel(excel_path, sheet_name='Total SANDAG')
        df = df[df['Veh_Tech'] == ' All Vehicles']
        # Extract the specific columns of interest
        working_df[(col1, 'ROG_TOTAL')] = df['ROG_TOTAL']
        working_df[(col1, 'NOx_TOTEX')] = df['NOx_TOTEX']

        
        # Append the extracted columns to the main DataFrame
        extracted_df = pd.concat([extracted_df, working_df], axis=1)
        
    return extracted_df
```
