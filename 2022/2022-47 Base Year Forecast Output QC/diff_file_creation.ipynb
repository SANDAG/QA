```python
emp_comparisons = pd.DataFrame()
emp_comparisons['mgra'] = DS41_2018['mgra']
emp_comparisons['DS41 2016'] = DS41_2016['emp_total']
emp_comparisons['DS41 2018'] = DS41_2018['emp_total']
emp_comparisons['DS44 2019'] = DS44_2019['emp_total']
emp_comparisons['DS41 2020'] = DS41_2020['emp_total']
emp_comparisons['2018 minus 2016'] = emp_comparisons['DS41 2018'] - emp_comparisons['DS41 2016']
emp_comparisons['2019 Minus 2018'] = emp_comparisons['DS44 2019'] - emp_comparisons['DS41 2018']
emp_comparisons['2020 Minus 2019'] = emp_comparisons['DS41 2020'] - emp_comparisons['DS44 2019']
emp_comparisons['2016 to 2018 PCT Change'] = ((emp_comparisons['DS41 2018']-emp_comparisons['DS41 2016']) / emp_comparisons['DS41 2016'] * 100).round(2)
emp_comparisons['2018 to 2019 PCT Change'] = ((emp_comparisons['DS44 2019']-emp_comparisons['DS41 2018']) / emp_comparisons['DS41 2018'] * 100).round(2)
emp_comparisons['2019 to 2020 PCT Change'] = ((emp_comparisons['DS41 2020']-emp_comparisons['DS44 2019']) / emp_comparisons['DS44 2019'] * 100).round(2)
```
