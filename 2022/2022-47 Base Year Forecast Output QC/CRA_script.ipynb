## Count of unique hhids by MGRA 
This is county number of persons 

```python
mgra_count = data[['mgra','hhid']].groupby(['mgra']).count()
mgra_count
```

## Count of unique hinccat1 by mgra 
This is count of number of houseold income types 

```python
# Unpiviting and concatonating what is needed
pivot = data[['mgra', 'hinccat1']].pivot(columns=['hinccat1'])

# This is a multicolumn dataframe, so I am removing that top column and only dealing with the data underneath
pivot = pivot['mgra']

pivot[pivot > 0] = 1 # The data in its current form has the hinccat1 value in the column then the MGRA in the row. I don't care about the specific MGRA here so I set all values greater than zero to one, this is so summing is easier. 

unique_hinccat1 = pd.concat([data[['mgra']], pivot], axis = 1) # Appending this new pivot to the MGRAs on the initial download
unique_hinccat1 = unique_hinccat1.groupby('mgra').sum() # Now summing up the unique values in each 
unique_hinccat1
```

# Count of hhid by untittype
This is by group quarter 

```python
# Unpiviting and concatonating what is needed
pivot = data[['hhid', 'unittype']].pivot(columns=['unittype'])

# This is a multicolumn dataframe, so I am removing that top column and only dealing with the data underneath
pivot = pivot['hhid']

pivot[pivot > 0] = 1 # The data in its current form has the hinccat1 value in the column then the MGRA in the row. I don't care about the specific MGRA here so I set all values greater than zero to one, this is so summing is easier. 

unique_hhid_by_unittype = pd.concat([data[['mgra']], pivot], axis = 1) # Appending this new pivot to the MGRAs on the initial download
unique_hhid_by_unittype = unique_hhid_by_unittype.groupby('mgra').sum() # Now summing up the unique values in each 
unique_hhid_by_unittype
```

## Sum of persons by MGRA 
Number of people in each MGRA

```python
sum_person_by_mgra = data[['mgra', 'persons']].groupby('mgra').sum()
sum_person_by_mgra
```
