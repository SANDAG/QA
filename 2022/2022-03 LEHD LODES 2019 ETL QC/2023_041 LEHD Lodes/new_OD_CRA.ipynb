# Loading Source Data (From R Drive)

```python
path = r'R:/DPOE/LEHD LODES/8.0/Source/od'
```

```python
# 2018 source data
csv = glob.glob(path + "/*2002.csv")
list = []

for filename in csv:
    df = pd.read_csv(filename, index_col=None, header=0, nrows=50)
    list.append(df)
    print(filename)
```

# Loading The Staging Table (SQL Server)

```python
conn = pyodbc.connect('Driver={ODBC Driver 17 for SQL Server};'
                      'Server=DDAMWSQL16.sandag.org;'
                      'Database=dpoe_stage;'
                      'Trusted_Connection=yes;')

#2018
query ="SELECT TOP(1000) * "\
                   "FROM [dpoe_stage].[lehd_lodes].[od_8_0]"\
                    "WHERE yr= 2002"
od_db = pd.read_sql_query(query, conn)
```

# Cleaning the source data

```python
od_csv = od_csv.drop(['createdate'], axis=1)
```

# Cleaning Staging Data

```python
od_db = od_db.drop(['type','yr'], axis=1)
```

```python
od_db['w_geocode'] = [i.lstrip('0') for i in od_db['w_geocode']]
```

```python
od_db['h_geocode'] = [i.lstrip('0') for i in od_db['h_geocode']]
```

# Filtering for just SD Data (Source Data)

```python
od_db = od_db[od_db['w_geocode'].astype(str).str.startswith('6073') | od_db['h_geocode'].astype(str).str.startswith('6073')]
od_db
```

# Filtering for just SD Data (Staging Data)

```python
od_csv = od_csv[od_csv['w_geocode'].astype(str).str.startswith('6073') | od_csv['h_geocode'].astype(str).str.startswith('6073')]
od_csv
```
