### DS42 == XPEF36

```python
%%time

# Import CSVs as dataframes in dictionary

yrs = ['2016', '2018', '2020', '2023', '2025', '2026', '2029', '2030', '2032', '2035', '2040', '2045', '2050']
df42 = {}

path = r'T:\socioec\Current_Projects\XPEF36\abm_csv\new_parking\New_mgra_based_input'
        
for x in yrs:
    file = path + '\mgra13_based_input' + x + '_05_np.csv'
    df42[x] = pd.read_csv(file)
    
    # Subset employment-related columns
    df42[x] = df42[x].iloc[:, df42[x].columns.str.startswith('emp_') | df42[x].columns.str.startswith('mgra')]
    
    # Add a year column to distinguish records when dataframes are combined
    df42[x]['yr'] = x
    
    # Check null values
    if df42[x].isnull().values.any():
        print(x + ' - ' + 'contains null values')
    else:
        print(x + ' - ' + 'pass')

df42['2016']
```

```python
# Concatenate all individual year dfs into a single df
df42_all = pd.concat(df42.values(), ignore_index=True)
df42_all
```

### DS38 == XPEF33

```python
%%time

# Import CSVs as dataframes in dictionary

yrs = ['2016', '2018', '2020', '2023', '2025', '2026', '2029', '2030', '2032', '2035', '2040', '2045', '2050']
df38 = {}

path = r'T:\socioec\Current_Projects\XPEF33\abm_csv\new_parking'
        
for x in yrs:
    file = path + '\mgra13_based_input' + x + '_05_np.csv'
    df38[x] = pd.read_csv(file)
    
    # Subset employment-related columns
    df38[x] = df38[x].iloc[:, df38[x].columns.str.startswith('emp_') | df38[x].columns.str.startswith('mgra')]
    
    # Add a year column to distinguish records when dataframes are combined
    df38[x]['yr'] = x

df38['2016']
```

```python
# Concatenate all individual year dfs into a single df
df38_all = pd.concat(df38.values(), ignore_index=True)
df38_all
```

# Aggregated MGRAs (i.e. region, jurisdiction, cpa)

## SQL mgra dim table

```python
# Connect to DDAMWSQL16 to access MGRA dimensions table
conn = pyodbc.connect('Driver={ODBC Driver 17 for SQL Server};'
                      'Server=DDAMWSQL16.sandag.org;'
                      'Database=socioec_data;'
                      'Trusted_Connection=yes;')

mgra = ['region','jurisdiction','cpa']
dim_mgra = {}

for x in mgra:
    qry = "SELECT DISTINCT \
        mgra, geotype, geozone \
        FROM demographic_warehouse.dim.mgra \
        WHERE series = 14 AND geotype = '" + x + "' \
        ORDER BY mgra"
    
    dim_mgra[x] = pd.read_sql_query(qry, conn)

dim_mgra['cpa']
```

## Summarize by mgra

```python
# Join df with mgra information
df42_mgra = {}
df38_mgra = {}

for x in mgra:
    df42_mgra[x] = pd.merge(df42_all, dim_mgra[x], on='mgra', how='left')
    df38_mgra[x] = pd.merge(df38_all, dim_mgra[x], on='mgra', how='left')

df42_mgra['jurisdiction']
```

```python
# Group & summarize by mgra and year
df42_grp = {}
df38_grp = {}

for x in mgra:
    df42_grp[x] = df42_mgra[x].groupby(['geozone','yr']).sum()
    df42_grp[x] = df42_grp[x].iloc[:, df42_grp[x].columns.str.startswith('emp_')]
    # df42_grp[x].loc['col total'] = df42_grp[x].sum(axis=0)
    # df42_grp[x].loc[:,'row total'] = df42_grp[x].sum(axis=1) - df42_grp[x]['emp_total']
    
    df38_grp[x] = df38_mgra[x].groupby(['geozone','yr']).sum()
    df38_grp[x] = df38_grp[x].iloc[:, df38_grp[x].columns.str.startswith('emp_')]
    # df38_grp[x].loc['col total'] = df38_grp[x].sum(axis=0)
    # df38_grp[x].loc[:,'row total'] = df38_grp[x].sum(axis=1) - df38_grp[x]['emp_total']

df42_grp['region']
```

## Compare 42 with 38

```python
df_comp = {}

for x in mgra:
    df_comp[x] = df42_grp[x].subtract(df38_grp[x])
    # df_comp[x].loc[:,'row-emp total diff'] = df_comp[x]['row total'] - df_comp[x]['emp_total']


df_comp['region']
```
