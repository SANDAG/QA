### DS41 == XPEF35

```python
%%time

# Import CSVs as dataframes in dictionary

yrs = ['2016', '2018', '2020', '2023', '2025', '2026', '2029', '2030', '2032', '2035', '2040', '2045', '2050']
df41 = {}

path = r'T:\socioec\Current_Projects\XPEF35\abm_csv\New_mgra_based_input'
        
for x in yrs:
    file = path + '\mgra13_based_input' + x + '_02.csv'
    df41[x] = pd.read_csv(file)
    
    # Subset employment-related columns
    df41[x] = df41[x].iloc[:, df41[x].columns.str.startswith('emp_') | df41[x].columns.str.startswith('mgra')]
    
    # Add a year column to distinguish records when dataframes are combined
    df41[x]['yr'] = x
    
    # Check null values
    if df41[x].isnull().values.any():
        print(x + ' - ' + 'contains null values')
    else:
        print(x + ' - ' + 'pass')

df41['2016']
```

```python
# Concatenate all individual year dfs into a single df
df41_all = pd.concat(df41.values(), ignore_index=True)
df41_all
```

### DS35 == XPEF31

```python
%%time

# Import CSVs as dataframes in dictionary

yrs = ['2016', '2018', '2020', '2023', '2025', '2026', '2029', '2030', '2032', '2035', '2040', '2045', '2050']
df35 = {}

path = r'T:\socioec\Current_Projects\XPEF31\abm_csv'
        
for x in yrs:
    file = path + '\mgra13_based_input' + x + '_01.csv'
    df35[x] = pd.read_csv(file)
    
    # Subset employment-related columns
    df35[x] = df35[x].iloc[:, df35[x].columns.str.startswith('emp_') | df35[x].columns.str.startswith('mgra')]
    
    # Add a year column to distinguish records when dataframes are combined
    df35[x]['yr'] = x

df35['2016']
```

```python
# Concatenate all individual year dfs into a single df
df35_all = pd.concat(df35.values(), ignore_index=True)
df35_all
```

# Aggregated MGRAs (i.e. region, jurisdiction, cpa)

## SQL mgra dim table

```python
# Connect to DDAMWSQL16 to access MGRA dimensions table
conn = pyodbc.connect('Driver={ODBC Driver 17 for SQL Server};'
                      'Server=DDAMWSQL16.sandag.org;'
                      'Database=socioec_data;'
                      'Trusted_Connection=yes;')

mgra = ['region','jurisdiction','cpa']
dim_mgra = {}

for x in mgra:
    qry = "SELECT DISTINCT \
        mgra, geotype, geozone \
        FROM demographic_warehouse.dim.mgra \
        WHERE series = 14 AND geotype = '" + x + "' \
        ORDER BY mgra"
    
    dim_mgra[x] = pd.read_sql_query(qry, conn)

dim_mgra['cpa']
```

## Summarize by mgra

```python
# Join df with mgra information
df41_mgra = {}
df35_mgra = {}

for x in mgra:
    df41_mgra[x] = pd.merge(df41_all, dim_mgra[x], on='mgra', how='left')
    df35_mgra[x] = pd.merge(df35_all, dim_mgra[x], on='mgra', how='left')

df41_mgra['jurisdiction']
```

```python
# Group & summarize by mgra and year
df41_grp = {}
df35_grp = {}

for x in mgra:
    df41_grp[x] = df41_mgra[x].groupby(['geozone','yr']).sum()
    df41_grp[x] = df41_grp[x].iloc[:, df41_grp[x].columns.str.startswith('emp_')]
    # df41_grp[x].loc['col total'] = df41_grp[x].sum(axis=0)
    # df41_grp[x].loc[:,'row total'] = df41_grp[x].sum(axis=1) - df41_grp[x]['emp_total']
    
    df35_grp[x] = df35_mgra[x].groupby(['geozone','yr']).sum()
    df35_grp[x] = df35_grp[x].iloc[:, df35_grp[x].columns.str.startswith('emp_')]
    # df35_grp[x].loc['col total'] = df35_grp[x].sum(axis=0)
    # df35_grp[x].loc[:,'row total'] = df35_grp[x].sum(axis=1) - df35_grp[x]['emp_total']

df41_grp['region']
```

## Compare 41 with 35

```python
df_comp = {}

for x in mgra:
    df_comp[x] = df41_grp[x].subtract(df35_grp[x])
    # df_comp[x].loc[:,'row-emp total diff'] = df_comp[x]['row total'] - df_comp[x]['emp_total']


df_comp['region']
```
