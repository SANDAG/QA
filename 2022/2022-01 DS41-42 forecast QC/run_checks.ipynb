# Comparison Function

```python
def find_difference(old_df_1, new_df_2, DSVAL):
    columns_to_check = ['hs', 'hs_sf', 'hs_mf', 'hs_mh', 'hh', 'hh_sf', 'hh_mf', 'hh_mh', 'gq_civ', 'gq_mil', 'hhs', 'pop', 'hhp']
    df = pd.DataFrame(columns=columns_to_check, index=old_df_1.index)
    for base in columns_to_check:
        new_column = new_df_2[base+'_'+DSVAL] - old_df_1[base+'_'+DSVAL]
        df[base] = new_column
    return df 
```

```python
# Diffs
ds35_diffs = find_difference(ds35_old, ds35_new, 'ds35')
ds38_diffs = find_difference(ds38_old, ds38_new, 'ds38')
ds41_diffs = find_difference(ds41_old, ds41_new, 'ds41')
ds42_diffs = find_difference(ds42_old, ds42_new, 'ds42')
```
