Perhaps a good way to begin classification is to identify arrest locations WITH numbers because numbers will be helpful in geocoding accurately.

```python
def has_numbers(inputString):
    if type(inputString) != str:
        return np.nan
    return any(char.isdigit() for char in inputString)
```

```python
ati_cleaning['arrest_location_numbers'] = ati_cleaning['LOCATION_OF_ARREST'].apply(has_numbers)
```

```python
no_numbers = ati_cleaning[ati_cleaning['arrest_location_numbers'] == False]
has_numbers = ati_cleaning[ati_cleaning['arrest_location_numbers'] == True]
```
