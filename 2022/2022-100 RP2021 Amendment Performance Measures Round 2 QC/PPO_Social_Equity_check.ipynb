```python
import sqlalchemy as sql
import pandas as pd
import numpy as np
```

```python
DDAM = sql.create_engine('mssql+pymssql://DDAMWSQL16/')
```

```python
query = f'''
SELECT [scenario_id]
      ,[measure]
      ,[metric]
      ,ROUND([value]*100,1) as [value]
      ,[updated_by]
      ,[updated_date]
  FROM [abm_14_2_0_reporting].[rp_2021].[results]
  WHERE
  (scenario_id = 767 OR scenario_id = 759 OR scenario_id = 764 OR scenario_id = 773 OR scenario_id = 766 OR scenario_id = 768)
  AND measure like 'M-5%'
  AND metric like '%Access Pct%' 
  AND metric not like 'Non-Senior%'  
  AND metric not like 'Population%'  
  AND metric not like 'Senior%'
 -- 767	759	764	773	759	766	768
'''
```

```python
server = pd.read_sql_query(query, con=DDAM)
server
```

```python
server = server[['scenario_id', 'measure', 'metric', 'value']]
server
```

```python
server = pd.pivot_table(server, index=['measure', 'metric'], columns = "scenario_id", values='value')
server
```

```python
server = server.reset_index()
```

```python
# ranking pattern:
# first: M-5a, M-5b...
# second: special rank for metric
# third: regional or mob hub
# fourth: 30 min or 45
```

```python
server['msr'] = server['measure'].str[:5]
```

```python
second_ranks = {
    'Low Income Access Pct - Transit - Speed One': 1,
    'Non-Low Income Access Pct - Transit - Speed One': 2,
    'Minority Access Pct - Transit - Speed One': 3,
    'Non-Minority Access Pct - Transit - Speed One': 4
}
```

```python
server['metric_rank'] = [second_ranks[metric] for metric in server['metric']]
```

```python
server['time'] = server['measure'].str[-10:]
```

```python
server = server.sort_values(by=['msr', 'metric_rank', 'measure', 'time'])
```
